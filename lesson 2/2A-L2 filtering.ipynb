## 2A - L2 Filtering

#### Setting up

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/dolphin_image.jpg")
```

```python
img.shape
```

```python
# resize image
img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
```

```python
# select a single channel
img = img[:, :, 1]
```

```python
plt.imshow(img, cmap="gray")
```

#### Gaussian Filters

Filters that represent the gaussian distribution

```python
def  create_gaussian_filter(x, y, sigma):
    return (1/(2 * np.pi * sigma ** 2)) * np.exp(-(x**2 + y**2)/sigma**2)

kernel_size = 30
x = np.linspace(-5, 5, kernel_size)
y = np.linspace(-5, 5, kernel_size)
x, y = np.meshgrid(x, y) # get 2D variables instead of 1D
```

##### Plotting the gaussian Surface

```python
kernel_size = 250
x = np.linspace(-5, 5, kernel_size)
y = np.linspace(-5, 5, kernel_size)
x, y = np.meshgrid(x, y) # get 2D variables instead of 1D
z = create_gaussian_filter(x, y, 2)

fig = plt.figure(figsize=(16, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface plot')
plt.show()
```

##### Visualizing gaussian filters

```python
kernel_size = 30
x = np.linspace(-5, 5, kernel_size)
y = np.linspace(-5, 5, kernel_size)
x, y = np.meshgrid(x, y) # get 2D variables instead of 1D

fig, ax = plt.subplots(2, 2, figsize=(16, 8))

ax[0, 0].imshow(create_gaussian_filter(x, y, 1), cmap="gray")
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 0].set_title("sigma=1")

ax[0, 1].imshow(create_gaussian_filter(x, y, 1.5), cmap="gray")
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[0, 1].set_title("sigma=1.5")

ax[1, 0].imshow(create_gaussian_filter(x, y, 2), cmap="gray")
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 0].set_title("sigma=2")

ax[1, 1].imshow(create_gaussian_filter(x, y, 3), cmap="gray")
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
ax[1, 1].set_title("sigma=3")
```

#### Denoising an Image using Gaussian Filters

##### Add noise to the img

```python
def create_gaussian_noise(mu, sigma, size):
    return np.array(mu + sigma * np.random.randn(size[0], size[1]))
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

size = img.shape

# original image
ax[0].imshow(img, cmap="gray")
ax[0].set_title("original Image")

# image with noise
noise = create_gaussian_noise(0, 10, size)
img_with_noise = img + noise
ax[1].imshow(img_with_noise, cmap="gray")
ax[1].set_title("img with noise")
```

##### Denoising with gaussian filters

```python
kernel_size = 30
x = np.linspace(-5, 5, kernel_size)
y = np.linspace(-5, 5, kernel_size)
x, y = np.meshgrid(x, y) # get 2D variables instead of 1D
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 8))

ax[0, 0].imshow(img, cmap="gray")
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 0].set_title("original img")

ax[0, 1].imshow(img_with_noise, cmap="gray")
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[0, 1].set_title("Image with noise")

kernel = create_gaussian_filter(x, y, 1)
out = cv2.filter2D(img_with_noise,cv2.CV_32F,kernel)/np.sum(kernel)
ax[1, 0].imshow(out, cmap="gray")
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 0].set_title("smoothing with kernel sigma 1")

kernel = create_gaussian_filter(x, y, 2)
out = cv2.filter2D(img_with_noise,cv2.CV_32F,kernel)/np.sum(kernel)
ax[1, 1].imshow(out, cmap="gray")
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
ax[1, 1].set_title("smoothing with kernel sigma 2")
```
