## 2A - L4 filters as templates

```python
import numpy as np
from scipy.signal import correlate, correlate2d
import matplotlib.pyplot as plt
```

### 1D Correlation and Template Matching

```python
signal = np.random.randint(-5, 5, 100)
template = signal[40:60] # choose a part of the signal as the template
```

```python
plt.plot(signal)
```

```python
# cross-correlation

template_size = len(template)

result = []
for i in range(0, len(signal), template_size):
    out = correlate(signal[i:i+template_size], template, "same")
    result.extend(out)
```

```python
plt.plot(result)
```

We can see a peak in the output where the template was taken from the signal

### 2D template matching

```python
test_img = np.ones((100, 100), dtype=np.uint8) * 255
template = np.ones((30, 30), dtype=np.uint8) * 10
test_img[50:80, 20:50] = template.copy()
```

```python
plt.imshow(test_img, cmap="gray")
```

```python
out = correlate2d(test_img, template, mode="same", boundary="symm")
```

```python
plt.imshow(out, cmap="gray")
```
