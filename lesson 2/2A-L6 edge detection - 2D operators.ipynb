## 2A-L6 Edge Detection: 2D Operators

```python
import cv2
import numpy as np
from scipy.signal import correlate2d
import matplotlib.pyplot as plt
```

### Gradient of Guassian Filters

##### Plot a regular gaussian filter

```python
def  create_gaussian_filter(x, y, sigma):
    return (1/(2 * np.pi * sigma ** 2)) * np.exp(-(x**2 + y**2)/sigma**2)
```

```python
kernel_size = 250
x = np.linspace(-5, 5, kernel_size)
y = np.linspace(-5, 5, kernel_size)
x, y = np.meshgrid(x, y) # get 2D variables instead of 1D
z = create_gaussian_filter(x, y, 2)

fig = plt.figure(figsize=(16, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Gaussian filter')
plt.show()
```

##### Take gradient of the gaussian filter

```python
kernel = np.array([[-1, 1]])
out = correlate2d(z, kernel, mode="same")
```

```python
fig = plt.figure(figsize=(16, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, out,cmap='viridis', edgecolor='none')
ax.set_title('Gradient of gaussian filter')
plt.show()
```
