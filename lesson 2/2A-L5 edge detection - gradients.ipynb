## 2A-L5 Edge Detection: Gradients

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
test_img = np.ones((500, 500), dtype=np.uint8) * 255
```

```python
test_img[200:300, 300:400] = 10
```

```python
plt.imshow(test_img, cmap="gray")
```

#### SobelX

```python
gx = cv2.Sobel(test_img,cv2.CV_64F,1,0,ksize=5)
```

```python
plt.imshow(gx, cmap="gray")
```

```python
# display the image and plot the intensity along a particular row

fig, ax = plt.subplots(2, 2, figsize=(16, 8))

# display the img
ax[0, 0].imshow(test_img, cmap="gray")
ax[0, 0].axhline(y = 250, color = 'r', linestyle = '-')

ax[0, 1].imshow(gx, cmap="gray")
ax[0, 1].axhline(y = 250, color = 'r', linestyle = '-')

# plot intensity along row 250
row_intensity = test_img[250, :]
ax[1, 0].plot(row_intensity)

row_intensity = gx[250, :]
ax[1, 1].plot(row_intensity)
```

#### SobelY

```python
gy = cv2.Sobel(test_img,cv2.CV_64F,0,1,ksize=5)
```

```python
plt.imshow(gy, cmap="gray")
```

#### Gradient magnitude

```python
g = np.sqrt(np.square(gx) + np.square(gy))
```

```python
plt.imshow(g, cmap="gray")
```

#### Gradient direction

```python
theta = np.arctan2(gy, gx)
```

```python
plt.imshow(theta, cmap="gray")
```
