## 2A - L1 Images as functions

### Load and display images

```python
from PIL import Image, ImageChops
import numpy as np
import matplotlib.pyplot as plt
```

```python
im1 = Image.open("data/cat_image.jpg")
im2 = Image.open("data/dolphin_image.jpg")
```

```python
# show image size
im1.size, im2.size
```

```python
# resize images so they have the same size
x = im1.size[0]//3
y = im1.size[1]//3
im1 = im1.resize((x, y))
im2 = im2.resize((x, y))
```

```python
im1
```

```python
im2
```

```python
# convert to numpy arrays
im1_arr = np.array(im1)
im2_arr = np.array(im2)
```

```python
# check data type
im1_arr.dtype
```

```python
# select a single channel
img1 = im1_arr[:, :, 1]
img2 = im2_arr[:, :, 2]
```

```python
# display the image and plot the intensity along a particular row

fig, ax = plt.subplots(1, 2, figsize=(16, 8))

# display the img
ax[0].imshow(img2, cmap="gray")
ax[0].axhline(y = 155, color = 'r', linestyle = '-')

# plot intensity along row 155
row_intensity = img2[155, :]
ax[1].plot(row_intensity)
```

we can see that the intensity goes down along the body of the dolphin as the color is black


### Operations on Images

#### Crop Image

```python
img_cropped = img2[90:200, 50: 350]
plt.imshow(img_cropped, cmap="gray")
```

#### Add two Images (Blending)

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

# 1. img1/2 + img2/2
ax[0].imshow(img1//2 + img2//2, cmap="gray")
ax[0].set_title("img1/2 + img2/2")

# 2. (img1 + img2)/2
ax[1].imshow((img1 + img2) // 2, cmap="gray")
ax[1].set_title("(img1 + img2) / 2")
```

#### Subtract two images

The problem with subtraction is that if (a - b) < 0 then it will be truncated to zero if we're using uint8 datatype, but what we want is the magnitude of (a - b) so for this we will use an inbuilt function provided by PIL

##### Normal Subtraction

```python
diff1 = img2 - img1
plt.imshow(diff1, cmap="gray")
plt.title("img2 - img1")
```

##### using library

```python
img1_img = Image.fromarray(img1)
img2_img = Image.fromarray(img2)
```

```python
diff2 = ImageChops.difference(img2_img, img1_img)
diff2
```

### Noise in Images

#### Gaussian Noise

##### Noise with varying std deviation

```python
def create_gaussian_noise(mu, sigma, size):
    return np.array(mu + sigma * np.random.randn(size[0], size[1]))
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 8))

mu= 128
size = img1.shape

ax[0, 0].imshow(create_gaussian_noise(mu, 2, size), cmap="gray")
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 0].set_title("mu: 128, sigma: 2")

ax[0, 1].imshow(create_gaussian_noise(mu, 32, size), cmap="gray")
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[0, 1].set_title("mu: 128, sigma: 32")

ax[1, 0].imshow(create_gaussian_noise(mu, 64, size), cmap="gray")
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 0].set_title("mu: 128, sigma: 64")

ax[1, 1].imshow(create_gaussian_noise(mu, 128, size), cmap="gray")
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
ax[1, 1].set_title("mu: 128, sigma: 128")
```

##### Adding noise to Image

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

# original image
ax[0].imshow(img1, cmap="gray")
ax[0].set_title("original Image")

# image with noise
noise = create_gaussian_noise(0, 8, size)
ax[1].imshow(img1 + noise, cmap="gray")
ax[1].set_title("img with noise")
```

##### Comparing the intensity along a row before and after adding noise

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 8))

size = img1.shape
noise = create_gaussian_noise(0, 8, size)

ax[0, 0].imshow(img1, cmap="gray")
ax[0, 0].axhline(y = 120, color = 'r', linestyle = '-')
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 0].set_title("original img")

ax[0, 1].imshow(noise + img1, cmap="gray")
ax[0, 1].axhline(y = 120, color = 'r', linestyle = '-')
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[0, 1].set_title("img with noise")

row_intensity = img1[120, :]
ax[1, 0].plot(row_intensity)
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 0].set_title("original intensity")

row_intensity = (img1 + noise)[120, :]
ax[1, 1].plot(row_intensity)
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
ax[1, 1].set_title("intensity with noise")
```
