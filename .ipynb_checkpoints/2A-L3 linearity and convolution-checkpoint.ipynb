#### Setting up

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/pattern_img.jpg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

#### Padding Methods

We'll use a gaussian filter to illustrate different padding methods

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 8))

out = cv2.GaussianBlur(img, ksize=(25, 25), sigmaX=2, sigmaY=2, borderType=cv2.BORDER_CONSTANT)
ax[0, 0].imshow(out, cmap="gray")
ax[0, 0].set_xticks([])
ax[0, 0].set_yticks([])
ax[0, 0].set_title("constant (clip filter)")

out = cv2.GaussianBlur(img, ksize=(25, 25), sigmaX=2, sigmaY=2, borderType=cv2.BORDER_WRAP)
ax[0, 1].imshow(out, cmap="gray")
ax[0, 1].set_xticks([])
ax[0, 1].set_yticks([])
ax[0, 1].set_title("wrap around")

out = cv2.GaussianBlur(img, ksize=(25, 25), sigmaX=2, sigmaY=2, borderType=cv2.BORDER_REFLECT)
ax[1, 0].imshow(out, cmap="gray")
ax[1, 0].set_xticks([])
ax[1, 0].set_yticks([])
ax[1, 0].set_title("reflect")

out = cv2.GaussianBlur(img, ksize=(25, 25), sigmaX=2, sigmaY=2, borderType=cv2.BORDER_REPLICATE)
ax[1, 1].imshow(out, cmap="gray")
ax[1, 1].set_xticks([])
ax[1, 1].set_yticks([])
ax[1, 1].set_title("replicate (copy edge)")
```

#### Sharpening Filter

```python
img = cv2.imread("data/cat_image.jpg")
```

```python
img = img[:, :, 1]
```

##### Create sharpening filter

```python
sharpening_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
```

```python
sharpening_kernel
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(img, cmap="gray")
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_title("original img")

out = cv2.filter2D(img, -1, sharpening_kernel)
ax[1].imshow(out, cmap="gray")
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title("Sharpened Image")
```

#### Median Filtering

```python
img = cv2.imread("data/cat_image.jpg")
```

```python
img = img[:, :, 1]
```

```python
# resize image
img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
```

##### Add salt and pepper noise

```python
x,y = img.shape
```

```python
salt = np.random.uniform(size=(x, y)) > 0.95
pepper = np.random.uniform(size=(x, y)) < 0.05
```

```python
np.sum(salt), np.sum(pepper)
```

```python
img[salt] = 255
img[pepper] = 0
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

ax[0].imshow(img, cmap="gray")
ax[0].set_title("Img with Salt and pepper noise")

out = cv2.medianBlur(img, 3)
ax[1].imshow(out, cmap="gray")
ax[1].set_title("Noisy img filtered with median blur")
```
